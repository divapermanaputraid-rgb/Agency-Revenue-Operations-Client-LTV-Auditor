# Phase 2: Data Integration & Metrics Engineering
**Objective:** Menggabungkan tabel dimensi dan fakta untuk membangun "Single Source of Truth". 
Pada tahap ini, kita akan menghitung metrik profitabilitas dasar yang akan menjadi penggerak utama animasi data di After Effects.

```python
import pandas as pd
import os

# Load raw data
dim_clients = pd.read_csv('../data/raw/dim_clients.csv')
fact_revenue = pd.read_csv('../data/raw/fact_revenue.csv')
fact_costs = pd.read_csv('../data/raw/fact_costs.csv')

# Validasi awal
print(f"Clients: {dim_clients.shape}")
print(f"Revenue Records: {fact_revenue.shape}")
print(f"Cost Records: {fact_costs.shape}")
```

```python
# Konversi kolom period ke datetime untuk analisis time-series
fact_revenue['period'] = pd.to_datetime(fact_revenue['period'])
fact_costs['period'] = pd.to_datetime(fact_costs['period'])

# cek apakah ada data duplikat
print(f"duplicate revenue: {fact_revenue.duplicated().sum()}")
print(f"duplicate costs: {fact_costs.duplicated().sum()}")
```

```python
#1. Gabungkan fact tabels (inner join karena setiap revenue harus punya biaya operasional)
fact_merged = pd.merge(
    fact_revenue,
    fact_costs,
    on=['period', 'client_id', 'service_line'],
    how='inner',
)
#2. gabungkan dengan dimension table (left join untuk mempertahankan semua data transaksi)
df_master = pd.merge(
    fact_merged,
    dim_clients,
    on='client_id',
    how='left'
)

print(f"Master Dataframe Shape: {df_master.shape}")
```
